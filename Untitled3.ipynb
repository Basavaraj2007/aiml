```python
import pandas as pd
mtcars=pd.read_csv('/content/mtcars.csv')
mtcars.head()

```

```python
import matplotlib.pyplot as plt
x=mtcars['mpg']
y=mtcars['wt']
plt.bar(x,y)
plt.title('frequency distribution')
plt.ylabel('frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=mtcars['mpg']
y=mtcars['wt']
plt.scatter(x,y)
plt.title('relationship between car weight and mpg')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
mtcars=pd.read_csv('/content/mtcars.csv')
plt.boxplot(mtcars['mpg'])
plt.title('Visualize the distribution of the mpg')
plt.ylabel('mpg')
plt.grid(True)
plt.show()
```
